## Wczytanie zbioru danych z użyciem biblioteki import.url

```python
import urllib.request  # the lib that handles the url stuff
url = "https://raw.githubusercontent.com/BioBarto/BioinformatykaRoslin2021/main/data_all.txt"
file = urllib.request.urlopen(url)

data_url = []
for line in urllib.request.urlopen(url):
    data_url.append(line.decode('utf-8'))

len(data_url)
```
